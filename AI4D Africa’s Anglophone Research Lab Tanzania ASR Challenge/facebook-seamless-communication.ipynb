###### Import relevant libraries

```python
import torch
import torchaudio
from seamless_communication.models.inference import Translator


# Initialize a Translator object with a multitask model, vocoder on the GPU.
translator = Translator("seamlessM4T_large", "vocoder_36langs", torch.device("cuda:0"), torch.float16)
```

###### Testing out the model

```python
# ASR
# This is equivalent to S2TT with `<tgt_lang>=<src_lang>`.
transcribed_text, _, _ = translator.predict("/kaggle/input/swahili-audio-dataset/all_wav/content/all_wav_files/audio_643a10c1.wav", "asr", "swh")
str(transcribed_text)
```

```python
import librosa
import IPython.display as ipd
import pandas as pd
```

```python
ipd.Audio(data="/kaggle/input/swahili-audio-dataset/all_wav/content/all_wav_files/audio_643a10c1.wav", autoplay=True, rate=16000)
```

###### Load in the dataset

```python
train = pd.read_csv("/kaggle/input/swahili-audio-dataset/train.csv")#.head(20)
test = pd.read_csv("/kaggle/input/swahili-audio-dataset/test.csv")#.head(20)
train["path"] = "/kaggle/input/swahili-audio-dataset/all_wav/content/all_wav_files/" + train["audio_ID"] +".wav"
test["path"] = "/kaggle/input/swahili-audio-dataset/all_wav/content/all_wav_files/" + test["audio_ID"] + ".wav"
test["sentence"] = " "
train
```

```python
test
```

###### Perform Inference on test set

```python
import tqdm
predictions = []
for i in tqdm.tqdm(test.path.to_list()):
    transcribed_text, _, _ = translator.predict(i, "asr", "swh")
    predictions.append(str(transcribed_text))
```

```python
test
```

```python
submission = pd.DataFrame(columns = ['audio_ID','pred_str'])

submission["pred_str"] = predictions
submission["audio_ID"] = test.audio_ID
submission.pred_str = submission["pred_str"].replace("", " ")
submission.head()
```

```python
submission.to_csv("Facebook Seamless communication.csv", index=False)
```
