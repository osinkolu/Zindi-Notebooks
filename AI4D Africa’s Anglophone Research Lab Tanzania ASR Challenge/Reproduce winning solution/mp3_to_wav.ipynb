```python
import pandas as pd
import os
```

```python
train = pd.read_csv("/content/drive/MyDrive/machine_learning_stories/Swahili ASR/train.csv")
test = pd.read_csv("/content/drive/MyDrive/machine_learning_stories/Swahili ASR/test.csv")
train["path"] = "/content/train_audios/" + train["path"]
test["path"] = "/content/test_audios/" + test["path"]
test["sentence"] = " "
train
```

```python
!mkdir "all_wav_files"
```

```python
import tqdm
from pydub import AudioSegment
for i in tqdm.tqdm(train.audio_ID):
 # print(i)
  #print("/content/train_wav/" + i + ".mp3")
  sound = AudioSegment.from_mp3("/content/train_audios/" + i + ".mp3")
  sound.export("/content/all_wav_files/" +i +".wav", format="wav")
```

```python
from pydub import AudioSegment
for i in tqdm.tqdm(test.audio_ID):
  sound = AudioSegment.from_mp3("/content/test_audios/" + i + ".mp3")
  sound.export("/content/all_wav_files/" +i +".wav", format="wav")
```
