```python
# Use a pipeline as a high-level helper
from transformers import pipeline
pipe = pipeline("automatic-speech-recognition", model="alamsher/wav2vec2-large-xlsr-53-common-voice-sw", device=0)
```

```python
from glob import glob
import pandas as pd
from tqdm import tqdm
```

```python
test = pd.read_csv("/content/drive/MyDrive/machine_learning_stories/Swahili ASR/test.csv")
test["path"] = "test_audios/" + test["path"]
test["sentence"] = " "
```

```python
test
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
res = []
for path in tqdm(test.path):
    res.append(pipe(path)['text'])
```

```python
test['sentence'] = res
```

```python
test[['audio_ID', 'sentence']].to_csv('alamsher on raw mp3 files.csv', index=False)
```
