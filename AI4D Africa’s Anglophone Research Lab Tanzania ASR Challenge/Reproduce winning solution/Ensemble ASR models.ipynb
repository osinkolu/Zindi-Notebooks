Ensemble both CSV files. The idea is simple. Check if the word spelling from the primary CSV file is correct, if it isn't check for the seccondary CSV file and use that if it is correct, else, keep results from the primary CSV file

```python
import pandas as pd

# Load the CSV files
primary_df = pd.read_csv("alamsher on raw mp3 files.csv")
secondary_df = pd.read_csv("Nemo 25 epochs yonas.csv")

# Load the Swahili dictionary list
swahili_dictionary_list = dictionary.to_list()

# Create a new list for ensembled text
ensembled_text = []

# Iterate through each row in the primary dataframe
for primary_row, secondary_row in zip(primary_df.itertuples(), secondary_df.itertuples()):
    primary_sentence = primary_row.sentence
    secondary_sentence = secondary_row.pred_str

    # Step 1: Check if the lengths of sentences are the same
    if len(primary_sentence.split()) != len(secondary_sentence.split()):
        ensembled_text.append(primary_sentence)  # Use primary sentence if lengths are not the same
    else:
        primary_words = primary_sentence.split()
        secondary_words = secondary_sentence.split()

        # Step 2: Check if all words in the primary sentence exist in the dictionary
        primary_words_in_dict = all(word in swahili_dictionary_list for word in primary_words)

        if primary_words_in_dict:
            ensembled_text.append(primary_sentence)  # Use primary sentence if all words are in dictionary
        else:
            # Check if all words in the secondary sentence exist in the dictionary
            secondary_words_in_dict = all(word in swahili_dictionary_list for word in secondary_words)

            if secondary_words_in_dict:
                ensembled_text.append(secondary_sentence)  # Use secondary sentence if all words are in dictionary
                print("picked", len(ensembled_text), "from Yonas")
    
            else:
                ensembled_text.append(primary_sentence)  # Use primary sentence if neither passes dictionary test

```
